**Univarite Analysis**




```python
df.shape
```

```python
df_setosa=df.loc[df['Species']=='setosa']
df_virgina=df.loc[df['Species']=='virginica']
df_versicolor=df.loc[df['Species']=='versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'o')
plt.plot(df_virgina['SepalLengthCm'],np.zeros_like(df_virgina['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'o')
plt.xlabel('Petal length')
plt.show()
```

#Bivariate Analysis

```python
# Change 'size' to 'height' to specify the height of each facet in inches.
# Adjust 'aspect' to control the width of each facet relative to its height.

sns.FacetGrid(df, hue='Species', height=5, aspect=1).map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm').add_legend()
plt.show()
```

**Multivariate Analysis**

```python
sns.pairplot(df,hue="Species",size=3)
```
